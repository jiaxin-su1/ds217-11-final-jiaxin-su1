### 1. `output/q1_data_info.txt`
**Format:** Plain text file
**Content:** Dataset information including:
- Dataset shape (rows × columns)
- Column names (one per line or comma-separated)
- Data types for each column
- Date range (start date and end date) - **REQUIRED if temporal data**
- Missing value counts for each column (column name: count)

**Example format:**
```
Dataset Shape: 50000 rows × 10 columns

Column Names:
- Measurement Timestamp
- Beach
- Water Temperature
- Air Temperature
...

Data Types:
- Measurement Timestamp: datetime64[ns]
- Beach: object
- Water Temperature: float64
...

Date Range:
Start: 2022-01-01 00:00:00
End: 2027-09-15 07:00:00

Missing Values:
- Water Temperature: 2500 (5.0%)
- Air Temperature: 1500 (3.0%)
...
```

````python
from IPython.display import Markdown

data_file = 'data/beach_sensors.csv'

if not os.path.exists(data_file):
    display(Markdown("### Data File Not Found"))
    display(Markdown("""
Please run `download_data.sh` to download the data:
```bash
chmod +x download_data.sh
./download_data.sh
```
"""))
    raise FileNotFoundError(f"Chicago Beach Weather Sensors data file not found: {data_file}. Run download_data.sh first.")

display(Markdown(f"Loading Chicago Beach Weather Sensors data from: `{data_file}`"))

# Load data - Parquet format (requires pyarrow, included in requirements.txt)
chi_sensors_df = pd.read_csv(data_file)
display(Markdown(f"**Loaded CSV file:** {len(chi_sensors_df):,} rows"))

# char(10) is newline
report = f"""
Dataset Shape: {chi_sensors_df.shape[0]} rows x {chi_sensors_df.shape[1]} columns

Columns Names:
{chr(10).join(f"- {col}" for col in chi_sensors_df.columns)}

Data Types:
{chr(10).join(f"- {col}: {dtype}" for col, dtype in chi_sensors_df.dtypes.items())}

Date Range:
Start: {chi_sensors_df["Measurement Timestamp"].min().replace("/", "-")}
End: {chi_sensors_df["Measurement Timestamp"].max().replace("/", "-")}

Missing Values:
{chr(10).join(f"- {col}: {chi_sensors_df[col].isna().sum()}" for col in chi_sensors_df.columns)}
"""

with open("output/q1_data_info.txt", "w") as f:
    # Your code here
    f.write(report)
````

### 2. `output/q1_exploration.csv`
**Format:** CSV file
**Required Columns (exact names):** `column_name`, `mean`, `std`, `min`, `max`, `missing_count`
**Content:** One row per numeric column in the dataset
- `column_name`: Name of the numeric column
- `mean`: Mean value (float)
- `std`: Standard deviation (float)
- `min`: Minimum value (float)
- `max`: Maximum value (float)
- `missing_count`: Number of missing values (integer)

**Example:**
```csv
column_name,mean,std,min,max,missing_count
Water Temperature,15.23,5.12,0.5,28.7,2500
Air Temperature,18.45,8.23,-5.2,35.8,1500
Wind Speed,6.78,4.56,0.1,25.3,0
```

````python
# Get describe() statistics
desc = chi_sensors_df.describe()

# Build your custom summary table
summary_df = pd.DataFrame({
    "column_name": desc.columns,
    "mean": desc.loc["mean"].round(2).values, # .loc to select by row label
    "std": desc.loc["std"].round(2).values,
    "min": desc.loc["min"].round(2).values,
    "max": desc.loc["max"].round(2).values,
    "missing_count": chi_sensors_df.select_dtypes(include='number').isna().sum().values
})

summary_df.to_csv("output/q1_exploration.csv", index=False)
````

### 3. `output/q1_visualizations.png`
**Format:** PNG image file
**Content:** At least 2 plots in a single figure (use subplots)
**Required plots:**
1. **Distribution plot:** Histogram or density plot of at least one numeric variable
2. **Time series plot:** Line plot showing a numeric variable over time (if temporal data)

**Requirements:**
- Clear axis labels (xlabel, ylabel)
- Title for each subplot
- Overall figure title (optional but recommended)
- Legend if multiple series shown
- Saved as PNG with sufficient resolution (dpi=150 or higher)

````python
chi_sensors_df["Measurement Timestamp"] = pd.to_datetime(
    chi_sensors_df["Measurement Timestamp"], errors="coerce"
)
chi_sensors_df = chi_sensors_df.sort_values("Measurement Timestamp")


fig, axes = plt.subplots(1, 2, figsize=(15, 12))
hist_data_air_T = chi_sensors_df['Air Temperature'].dropna()
axes[0].hist(hist_data_air_T, bins=30,color = "skyblue")
axes[0].set_title('Histogram for Air Temperature')
axes[0].set_xlabel("Air Temperature (°F)")
axes[0].set_ylabel("frequency")
axes[0].grid(True)


axes[1].plot(chi_sensors_df["Measurement Timestamp"],chi_sensors_df["Air Temperature"],linewidth=1)

axes[1].set_title(f"Air Temperature Over Time")
axes[1].set_xlabel("Time")
axes[1].set_ylabel("Air Temperature")

plt.savefig("output/q1_visualizations.png", dpi=300, bbox_inches="tight")
plt.show()
````
